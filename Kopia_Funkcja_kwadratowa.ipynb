```
a = int(input("podja liczbę dodatnią dla której funkcja ma zacząć rysować wykres i skończyć na przeciwnej"))

import matplotlib.pyplot as plt
import numpy as np

# Data for plotting
x = np.arange(-a , a , 0.01)
y = x*x

fig, ax = plt.subplots()
ax.plot(x, y)

ax.set(Xlabel='x', ylabel='y',
       title='$funkcja.kwadratowa$')
ax.grid()
plt.show()
```
